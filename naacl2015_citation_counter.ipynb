This IPython notebook processes "sentences" in the NAACL 2015 Proceedings and counts the most common sentences, which tend to be highly cited papers. 

This is just for fun, and might not be completely correct. Feel free to fork the code. More practically, I thought it'd be useful to see what background papers are most worth reading to understand where the NAACL community is at. 

```python
import os
txt_dir = './split/pgs/txt/'
files = os.listdir( './split/pgs/txt/' )

sentences = []
for fname in files:
    fpath = os.path.join( txt_dir, fname )
    with open( fpath ) as f:
        x = f.read().split('.')
        
        sentences += x

enum_sentences = list( enumerate( sentences ) )
sent_enum_dict = dict( [ (i,idx) for (idx,i) in enum_sentences ])
print len( sentences )
```
